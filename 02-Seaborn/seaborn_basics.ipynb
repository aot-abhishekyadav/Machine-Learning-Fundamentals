```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
'''
Correlation with Heatmap
A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection
'''
```

```python
df1 = df[['total_bill','tip','size']] # Select only numerical columns before calculating the correlation
df1.corr()
```

```python
sns.heatmap(df1.corr())
```

```python
'''
joinPlot
A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot
'''
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df1, kind = 'hex')
# Hex is use for giving shape of hexagon
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df1, kind = 'reg')
```

```python
'''
Pair plot
A “pairs plot” is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables. we use pairplot when we have more than two independent variable
'''
```

```python
sns.pairplot(df1)
```

```python
sns.pairplot(df, hue = 'sex') # we use df because in df we have sex in df1 we don't have sex
```

```python
sns.pairplot(df, hue = 'smoker')
```

```python
df['sex'].value_counts()
```

```python
'''
Categorical Plots
Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about
1.boxplot
2.violinplot
3.countplot
4.bar plot
'''
```

```python
# countplot
sns.countplot(x = 'sex', data = df)
```

```python
sns.countplot(y = 'smoker', data = df) # in this we can only pass value to either x or y we cannot pass value for both x and y it gives error
```

```python
# Barplot
# in this we can pass value for both x and y
sns.barplot(x = 'sex', y = 'total_bill', data = df)
```

```python

sns.barplot(x = 'day', y = 'total_bill', data = df)
```

```python
sns.barplot(x = 'day', y = 'tip', data = df)
```

```python
# Boxplot
```

```python
sns.boxplot(x = 'sex', y = 'total_bill', data = df)
```

```python
sns.boxplot(x = 'smoker', y = 'total_bill', data = df)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = df, palette = 'rainbow') # palette is use for giving colour to box
```

```python
sns.boxplot(data = df , orient = 'v') # orient make Boxplot of all the numerical value in the dataset
```

```python
sns.boxplot(x = 'day', y = 'total_bill', hue = 'smoker', data = df)
```

```python
sns.boxplot(x = 'day', y = 'tip', hue = 'sex', data = df)
```

```python
# violin plot
# Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot it is the mixture of both kernal and boxplot
```

```python
sns.violinplot(x = 'total_bill', y = 'day', data = df)
```

```python
sns.violinplot(x = 'total_bill', y = 'day', hue = 'sex', data = df, palette = 'rainbow')
```
