```python
# Converting Dataframe into json
df = pd.DataFrame([['a' , 'b'] , ['c' , 'd']],
index = ['row1' , 'row2'],
columns = ['col1' , 'col2'])
df
```

```python
df.to_json() # Converting into json
```

```python
df.to_json(orient = 'index') # High Priority is given to row
```

```python
df.to_json(orient = 'columns') # Print json like deafult mode high priority
                               # is given to columns
```

```python
df.to_json(orient = 'records') # High priority is given to columns and it ignores the row
```

```python
df.to_json(orient = 'split') # yai sab kaa alag alag data print karta hai
```

```python
df.to_json(orient = 'table')
```

```python
data = [
    {
        "id": 1,
        "name": "Cole Volk",
        "fitness": {"height": 130, "weight": 60},
    },
    {"name": "Mark Reg", "fitness": {"height": 130, "weight": 60}},
    {
        "id": 2,
        "name": "Faye Raker",
        "fitness": {"height": 130, "weight": 60},
    },
]
```

```python
pd.json_normalize(data)
```

```python
pd.json_normalize(data, max_level = 0)
# Max level 0 pai vo nested list ko alg column mai separate nhi karta hai
```

```python
pd.json_normalize(data, max_level = 1)
```

```python
data = [
    {
        "state": "Florida",
        "shortname": "FL",
        "info": {"governor": "Rick Scott"},
        "counties": [
            {"name": "Dade", "population": 12345},
            {"name": "Broward", "population": 40000},
            {"name": "Palm Beach", "population": 60000},
        ],
    },
    {
        "state": "Ohio",
        "shortname": "OH",
        "info": {"governor": "John Kasich"},
        "counties": [
            {"name": "Summit", "population": 1234},
            {"name": "Cuyahoga", "population": 1337},
        ],
    },
]
```

```python
pd.json_normalize(data)
```

```python
pd.json_normalize(data, max_level = 0)
```

```python
# using of record parameter
pd.json_normalize(data,"counties")
# counties ko print karne mai mushkil ho rhi thi isliye humne usse pehele alg sai print kara liya
```

```python
# using of meta parameter
pd.json_normalize(data,"counties" , ["state" , "shortname" , ["info" , "governor"]])
# meta ki help sai humne itne hard data ko easy way sai seprate kar kai print kar diya
```
